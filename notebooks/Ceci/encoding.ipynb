```python
import random
import pandas as pd

# Define the mapping for the 'occupation' column
occupation_mapping = {
    'Scientist': 1,
    'Teacher': 2,
    'Engineer': 3,
    'Entrepreneur': 4,
    'Developer': 5,
    'Lawyer': 6,
    'Media Manager': 7,
    'Doctor': 8,
    'Journalist': 9,
    'Manager': 10,
    'Accountant': 11,
    'Musician': 12,
    'Mechanic': 13,
    'Writer': 14,
    'Architect': 15
}

# Define the mapping for the 'credit_mix' column
credit_mix_mapping = {
    'Good': 1,
    'Standard': 2,
    'Bad': 3
}

# Define the mapping for the 'payment_of_min_amount' column
payment_of_min_amount_mapping = {
    'No': 1,
    'NM': 2,
    'Yes': 3
}

# Define the mapping for the 'payment_behaviour' column
payment_behaviour_mapping = {
    'High_spent_Small_value_payments': 1,
    'Low_spent_Large_value_payments': 2,
    'Low_spent_Medium_value_payments': 3,
    'Low_spent_Small_value_payments': 4,
    'High_spent_Medium_value_payments': 5,
    'High_spent_Large_value_payments': 6
}

# Define the mapping for the 'credit_score' column
credit_score_mapping = {
    'Good': 0,
    'Standard': 1,
    'Poor': 2
}

# Define exact keywords that indicate 'not specified' or similar values, including "No Data"
not_specified_keywords = ['Not Specified', 'Unknown', 'Other', 'N/A', 'Not Available', 'Unspecified', 'No Data']

# Function to apply encoding based on a given mapping
def encode_column(df, column_name, mapping_dict, not_specified_keywords=None):
    """
    Encodes a specific column in the DataFrame using the provided mapping dictionary.
    This will replace the original values in the column with the encoded values.

    Parameters:
    df (pandas.DataFrame): The DataFrame containing the column to encode.
    column_name (str): The name of the column to encode.
    mapping_dict (dict): The mapping dictionary that defines the encoding for the column.
    not_specified_keywords (list, optional): List of keywords that should be replaced with a specific value (e.g., 666).

    Returns:
    df (pandas.DataFrame): The DataFrame with the encoded column.
    """
    if not_specified_keywords:
        # Replace the not specified values with 666
        df[column_name] = df[column_name].apply(lambda x: 666 if x in not_specified_keywords else x)

    # Map the values using the provided mapping dictionary
    df[column_name] = df[column_name].map(mapping_dict).fillna(df[column_name])
    return df


# Sample DataFrame (replace this with your actual DataFrame)
# df = pd.read_csv('your_data.csv')  # Replace with your actual DataFrame loading step

# Example usage of encoding for each column
# Encoding 'occupation' column
df = encode_column(df, 'occupation', occupation_mapping)

# Encoding 'credit_mix' column
df = encode_column(df, 'credit_mix', credit_mix_mapping)

# Encoding 'payment_of_min_amount' column
df = encode_column(df, 'payment_of_min_amount', payment_of_min_amount_mapping)

# Encoding 'payment_behaviour' column
df = encode_column(df, 'payment_behaviour', payment_behaviour_mapping)

# Encoding 'credit_score' column
df = encode_column(df, 'credit_score', credit_score_mapping)

# Encoding 'type_of_loan' column with special handling for 'not specified' values
loan_mapping = {}
unique_loans = df['type_of_loan'].unique()
for loan in unique_loans:
    if loan in not_specified_keywords:
        loan_mapping[loan] = 666  # Assign code 666 for exact matches
    else:
        loan_mapping[loan] = random.randint(100, 999)  # Assign a random 3-digit code to other values

df = encode_column(df, 'type_of_loan', loan_mapping, not_specified_keywords)

# To verify, you can print the first few rows of the updated DataFrame
df.head()
```
