```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
url = 'https://raw.githubusercontent.com/alvarofavale/week7_ml/refs/heads/main/data/encoded/encoded_data.csv'
df = pd.read_csv(url)

# Set display option to show all columns
pd.set_option('display.max_columns', None)

df.head()
```

```python
# Define features (X) and target (y) for classification
# Assuming 'credit_score' is the target column (modify if needed)
features = df.drop(columns=["credit_score", 'name'])  # Drop the target column from features
target = df["credit_score"]  # Use the 'credit_score' column as the target

# Split the data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
# Create instances of the normalizer and scaler
normalizer = MinMaxScaler()
scaler = StandardScaler()

# Normalize the training and test data using MinMaxScaler
X_train_norm_np = normalizer.fit_transform(X_train)
X_test_norm_np = normalizer.transform(X_test)

# Convert the normalized data back to DataFrame with the original columns and index
X_train_norm_df = pd.DataFrame(X_train_norm_np, columns=X_train.columns, index=X_train.index)
X_test_norm_df = pd.DataFrame(X_test_norm_np, columns=X_test.columns, index=X_test.index)

# Standardize the training and test data using StandardScaler
X_train_standarized_np = scaler.fit_transform(X_train)
X_test_standarized_np = scaler.transform(X_test)

# Convert the standardized data back to DataFrame with the original columns and index
X_train_standarized_df = pd.DataFrame(X_train_standarized_np, columns=X_train.columns, index=X_train.index)
X_test_standarized_df = pd.DataFrame(X_test_standarized_np, columns=X_test.columns, index=X_test.index)

# Display the first few rows of the normalized and standardized DataFrames as an example
print("Normalized Training Data:")
X_train_norm_df.head()
```

```python
print("\nNormalized Testing Data:")
X_test_norm_df.head()
```

```python
print("\nStandardized Training Data:")
X_train_standarized_df.head()
```

```python
print("\nStandardized Testing Data:")
X_test_standarized_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Create instances of the LinearRegression model
lin_reg = LinearRegression()
lin_reg2 = LinearRegression()

# Fit the model to the normalized training data
lin_reg.fit(X_train_norm_df, y_train)  # Determines the b0 and b1's values

# Fit the model to the standardized training data
lin_reg2.fit(X_train_standarized_df, y_train)

# Make predictions on the test data (normalized)
y_pred_test = lin_reg.predict(X_test_norm_df)

# Calculate and print performance metrics for the normalized data
print(f"MAE (Normalized): {mean_absolute_error(y_pred_test, y_test): .2f}")  # Mean Absolute Error
print(f"RMSE (Normalized): {mean_squared_error(y_pred_test, y_test, squared=False): .2f}")  # Root Mean Squared Error
print(f"R2 score (Normalized): {lin_reg.score(X_test_norm_df, y_test): .2f}")  # R-squared Score

# Make predictions on the test data (standardized)
y_pred_test2 = lin_reg2.predict(X_test_standarized_df)

# Calculate and print performance metrics for the standardized data
print(f"MAE (Standardized): {mean_absolute_error(y_pred_test2, y_test): .2f}")  # Mean Absolute Error
print(f"RMSE (Standardized): {mean_squared_error(y_pred_test2, y_test, squared=False): .2f}")  # Root Mean Squared Error
print(f"R2 score (Standardized): {lin_reg2.score(X_test_standarized_df, y_test): .2f}")  # R-squared Score

# Define tolerance level (e.g., 0.1)
tolerance = 0.1

# Calculate accuracy for the normalized predictions
accuracy_norm = (abs(y_pred_test - y_test) <= tolerance).mean() * 100

# Calculate accuracy for the standardized predictions
accuracy_std = (abs(y_pred_test2 - y_test) <= tolerance).mean() * 100

# Print the accuracy for both normalized and standardized predictions
print(f"Accuracy (Normalized) within tolerance {tolerance}: {accuracy_norm:.2f}%")
print(f"Accuracy (Standardized) within tolerance {tolerance}: {accuracy_std:.2f}%")
```

```python
# Get the coefficients for the linear regression model (trained on normalized data)
lin_reg_coef = {feature: coef for feature, coef in zip(X_train_norm_df.columns, lin_reg.coef_)}

# Convert the coefficients dictionary into a DataFrame for easier presentation
coef_df = pd.DataFrame(list(lin_reg_coef.items()), columns=['Feature', 'Coefficient'])

# Add the intercept to the DataFrame
intercept_df = pd.DataFrame({'Feature': ['Intercept'], 'Coefficient': [lin_reg.intercept_]})

# Combine the intercept with the coefficients DataFrame
coef_df = pd.concat([coef_df, intercept_df], ignore_index=True)

# Display the coefficients and intercept in a table format
print(coef_df)
```
