```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
url = 'https://raw.githubusercontent.com/alvarofavale/week7_ml/refs/heads/main/data/encoded/encoded_data.csv'
df = pd.read_csv(url)

# Set display option to show all columns
pd.set_option('display.max_columns', None)

df.head()
```

```python
# Define features (X) and target (y) for classification
# Assuming 'credit_score' is the target column (modify if needed)
features = df.drop(columns=["credit_score", 'name'])  # Drop the target column from features
target = df["credit_score"]  # Use the 'credit_score' column as the target

# Split the data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
# Create an instance of the MinMaxScaler
normalizer = MinMaxScaler()

# Example: Fit and transform the features in your dataset
# Assuming 'X_train' is your training data (with numerical features)
X_train_scaled = normalizer.fit_transform(X_train)

# For testing data, you should only transform (don't fit again)
X_test_scaled = normalizer.transform(X_test)

# Print the scaled training data as an example
X_train.head()
```

```python
# Create and train the KNN Classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred_class = knn_classifier.predict(X_test_scaled)

# Evaluate the classifier's performance
accuracy = accuracy_score(y_test, y_pred_class)
print(f"KNN Classifier Accuracy: {accuracy * 100:.2f}%")

# You can also use classification report for more detailed evaluation
print("Classification Report:")
print(classification_report(y_test, y_pred_class))
```

```python
# Create and train the KNN Regressor
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred_reg = knn_regressor.predict(X_test_scaled)

# Evaluate the regressor's performance
mse = mean_squared_error(y_test, y_pred_reg)
r2 = r2_score(y_test, y_pred_reg)
print(f"KNN Regressor - Mean Squared Error: {mse}")
print(f"KNN Regressor - R-squared: {r2}")
```
