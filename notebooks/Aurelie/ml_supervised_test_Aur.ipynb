```python
# "Cleaning"
X = df.drop(columns=['Credit_Score', 'ID', 'Customer_ID', 'Name', 'SSN', "Occupation", "Type_of_Loan"])
y = df['Credit_Score']

# https://scikit-learn.org/dev/modules/generated/sklearn.preprocessing.LabelEncoder.html
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Encode categorical var 
X = pd.get_dummies(X)

# train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

#
```

 ### Linear 

```python
normalizer = MinMaxScaler()

normalizer.fit(X_train)
```

```python
scaler = StandardScaler()

scaler.fit(X_train)
```

```python
X_train_norm_np = normalizer.transform(X_train)

X_test_norm_np = normalizer.transform(X_test)

print(X_test)
```

```python
X_train_norm_df = pd.DataFrame(X_train_norm_np, columns = X_train.columns, index=X_train.index)
X_test_norm_df = pd.DataFrame(X_test_norm_np, columns = X_test.columns, index=X_test.index)
```

```python
X_train_standarized_np = scaler.transform(X_train)
X_test_standarized_np = scaler.transform(X_test)

X_train_standarized_df = pd.DataFrame(X_train_standarized_np, columns = X_train.columns, index=X_train.index)
X_test_standarized_df  = pd.DataFrame(X_test_standarized_np, columns = X_test.columns, index=X_test.index)
```

```python
# linearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_pred_test = lin_reg.predict(X_test_norm_df)

#lin_reg.score(X_test_norm_df, y_test)
print("Linear Regression")
print(f"MAE {mean_absolute_error(y_pred_test, y_test): .2f}") # mean(abs(error)) = mean(abs(y_test - y_pred_test))
print(f"MSE, {mean_squared_error(y_pred_test, y_test): .2f}") # mean(err **2) = mean( (y_test - y_pred_test)^2 ) ) # b0, b1, b2... 
print(f"RMSE, {root_mean_squared_error(y_pred_test, y_test): .2f}") # sqrt( mean( (y_test - y_pred_test)^2 ) ) # b0, b1, b2...
print(f"R2 score, {lin_reg.score(X_test_norm_df, y_test): .2f}")
```

### Random Forest

```python
# Random Forest Regressor
forest = RandomForestRegressor(n_estimators=100, max_depth=20)
forest.fit(X_train, y_train)
```

```python
# Prediction
y_pred_test_rf = forest.predict(X_test)
```

```python
# Evaluate
print("random forest regressor")
print(f"MAE: {mean_absolute_error(y_test, y_pred_test_rf): .2f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_test_rf): .2f}")
print(f"RMSE: {root_mean_squared_error(y_test, y_pred_test_rf): .2f}")
print(f"R2 score: {forest.score(X_test, y_test): .2f}")
```

```python
forest_class = RandomForestClassifier(n_estimators=100, max_depth=20)
forest_class.fit(X_train, y_train)
```

```python
# Prediction
y_pred_test_rf = forest_class.predict(X_test)
```

```python
print("random forest classifier")
print(f"MAE: {mean_absolute_error(y_test, y_pred_test_rf): .2f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_test_rf): .2f}")
print(f"RMSE: {root_mean_squared_error(y_test, y_pred_test_rf): .2f}")
print(f"R2 score: {forest_class.score(X_test, y_test): .2f}")
```
